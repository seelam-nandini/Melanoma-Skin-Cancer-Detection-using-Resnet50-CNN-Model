```python
import os
import glob
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from tensorflow.keras.callbacks import EarlyStopping,Callback
from tensorflow.keras.applications import ResNet50, MobileNetV2, InceptionV3, VGG16
from tensorflow.keras.applications.resnet50 import preprocess_input as preprocess_input_resnet50
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input as preprocess_input_mobilenetv2
from tensorflow.keras.applications.inception_v3 import preprocess_input as preprocess_input_inceptionv3
from tensorflow.keras.applications.vgg16 import preprocess_input as preprocess_input_vgg16
from sklearn.metrics import classification_report
```

```python
file_path = 'traindata'
```

```python
name_class = os.listdir(file_path)
name_class
```

```python
file_paths = list(glob.glob(file_path + '/**/*.*', recursive=True))
```

```python
labels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1], file_paths))
```

```python
filepath = pd.Series(file_paths, name='Filepath').astype(str)
labels = pd.Series(labels, name='label')
data = pd.concat([filepath, labels], axis=1)
data = data.sample(frac=1).reset_index(drop=True)
data.head(3)
```

```python
counts = data['label'].value_counts()
# Plotting the bar plot using Seaborn
sns.barplot(x=counts.index, y=counts)
plt.xlabel('Type')
plt.xticks(rotation=90)
plt.show()
```

```python
train, test = train_test_split(data, test_size=0.25, random_state=42)
```

```python
fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(10,8), subplot_kw={'xticks': [], 'yticks': []})
for i, ax in enumerate(axes.flat):
    ax.imshow(plt.imread(data.Filepath[i]))
    ax.set_title(data.label[i])
plt.tight_layout()
plt.show()
```

```python
train_datagen_resnet50 = ImageDataGenerator(preprocessing_function=preprocess_input_resnet50)
train_datagen_mobilenetv2 = ImageDataGenerator(preprocessing_function=preprocess_input_mobilenetv2)
train_datagen_inceptionv3 = ImageDataGenerator(preprocessing_function=preprocess_input_inceptionv3)
train_datagen_vgg16 = ImageDataGenerator(preprocessing_function=preprocess_input_vgg16)

test_datagen_resnet50 = ImageDataGenerator(preprocessing_function=preprocess_input_resnet50)
test_datagen_mobilenetv2 = ImageDataGenerator(preprocessing_function=preprocess_input_mobilenetv2)
test_datagen_inceptionv3 = ImageDataGenerator(preprocessing_function=preprocess_input_inceptionv3)
test_datagen_vgg16 = ImageDataGenerator(preprocessing_function=preprocess_input_vgg16)
```

```python
# Defining train, validation, and test generators for each model
# ResNet50
train_gen_resnet50 = train_datagen_resnet50.flow_from_dataframe(
    dataframe=train,
    x_col='Filepath',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=32,
    shuffle=True,
    seed=42
)
valid_gen_resnet50 = train_datagen_resnet50.flow_from_dataframe(
    dataframe=test,
    x_col='Filepath',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=32,
    shuffle=False,
    seed=42
)
test_gen_resnet50 = test_datagen_resnet50.flow_from_dataframe(
    dataframe=test,
    x_col='Filepath',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=32,
    shuffle=False
)
# MobileNetV2
train_gen_mobilenetv2 = train_datagen_mobilenetv2.flow_from_dataframe(
    dataframe=train,
    x_col='Filepath',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=64,
    shuffle=True,
    seed=42
)
valid_gen_mobilenetv2 = train_datagen_mobilenetv2.flow_from_dataframe(
    dataframe=test,
    x_col='Filepath',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=64,
    shuffle=False,
    seed=42
)
test_gen_mobilenetv2 = test_datagen_mobilenetv2.flow_from_dataframe(
    dataframe=test,
    x_col='Filepath',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=64,
    shuffle=False
)
# InceptionV3
train_gen_inceptionv3 = train_datagen_inceptionv3.flow_from_dataframe(
    dataframe=train,
    x_col='Filepath',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=64,
    shuffle=True,
    seed=42
)
valid_gen_inceptionv3 = train_datagen_inceptionv3.flow_from_dataframe(
    dataframe=test,
    x_col='Filepath',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=64,
    shuffle=False,
    seed=42
)
test_gen_inceptionv3 = test_datagen_inceptionv3.flow_from_dataframe(
    dataframe=test,
    x_col='Filepath',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=64,
    shuffle=False
)
# VGG16
train_gen_vgg16 = train_datagen_vgg16.flow_from_dataframe(
    dataframe=train,
    x_col='Filepath',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=64,
    shuffle=True,
    seed=42
)
valid_gen_vgg16 = train_datagen_vgg16.flow_from_dataframe(
    dataframe=test,
    x_col='Filepath',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=64,
    shuffle=False,
    seed=42
)
test_gen_vgg16 = test_datagen_vgg16.flow_from_dataframe(
    dataframe=test,
    x_col='Filepath',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=64,
    shuffle=False
)
```

```python
# Defining pretrained models
pretrained_model_resnet50 = ResNet50(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet',
    pooling='avg'
)
pretrained_model_mobilenetv2 = MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet',
    pooling='avg'
)
pretrained_model_inceptionv3 = InceptionV3(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet',
    pooling='avg'
)
pretrained_model_vgg16 = VGG16(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet',
    pooling='avg'
)
```

```python
# Freezing pretrained models
pretrained_model_resnet50.trainable = False
pretrained_model_mobilenetv2.trainable = False
pretrained_model_inceptionv3.trainable = False
pretrained_model_vgg16.trainable = False
```

```python
# ResNet50
inputs_resnet50 = pretrained_model_resnet50.input
x_resnet50 = Dense(128, activation='relu')(pretrained_model_resnet50.output)
x_resnet50 = Dense(128, activation='relu')(x_resnet50)

# MobileNetV2
inputs_mobilenetv2 = pretrained_model_mobilenetv2.input
x_mobilenetv2 = Dense(128, activation='relu')(pretrained_model_mobilenetv2.output)
x_mobilenetv2 = Dense(128, activation='relu')(x_mobilenetv2)

# InceptionV3
inputs_inceptionv3 = pretrained_model_inceptionv3.input
x_inceptionv3 = Dense(128, activation='relu')(pretrained_model_inceptionv3.output)
x_inceptionv3 = Dense(128, activation='relu')(x_inceptionv3)

# VGG16
inputs_vgg16 = pretrained_model_vgg16.input
x_vgg16 = Dense(128, activation='relu')(pretrained_model_vgg16.output)
x_vgg16 = Dense(128, activation='relu')(x_vgg16)
```

```python
# Defining output layers for each model
outputs_resnet50 = Dense(2, activation='softmax')(pretrained_model_resnet50.output)
outputs_mobilenetv2 = Dense(2, activation='softmax')(pretrained_model_mobilenetv2.output)
outputs_inceptionv3 = Dense(2, activation='softmax')(pretrained_model_inceptionv3.output)
outputs_vgg16 = Dense(2, activation='softmax')(pretrained_model_vgg16.output)
```

```python
# Creating models
model_resnet50 = Model(inputs=pretrained_model_resnet50.input, outputs=outputs_resnet50)
model_mobilenetv2 = Model(inputs=pretrained_model_mobilenetv2.input, outputs=outputs_mobilenetv2)
model_inceptionv3 = Model(inputs=pretrained_model_inceptionv3.input, outputs=outputs_inceptionv3)
model_vgg16 = Model(inputs=pretrained_model_vgg16.input, outputs=outputs_vgg16)
```

```python
# Compiling models
model_resnet50.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy'] 
)
model_mobilenetv2.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy'] 
)
model_inceptionv3.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy'] 
)
model_vgg16.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy'] 
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(
    monitor='val_loss',  
    patience=5,    
    restore_best_weights=True
)
print("\033[1mTraining ResNet50 Model\033[0m")
history_resnet50 = model_resnet50.fit(
    train_gen_resnet50,
    validation_data=valid_gen_resnet50,
    epochs=50,
    callbacks=[early_stopping] 
)
```

```python
model_resnet50.save("resnet50_melanomamodel.h5")
import matplotlib.pyplot as plt

epochs = range(1, len(history_resnet50.history['accuracy']) + 1)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

# Plotting accuracy and validation accuracy for ResNet50 model on the first subplot
ax1.plot(epochs, history_resnet50.history['accuracy'], label='ResNet50 Accuracy', color='green')
ax1.plot(epochs, history_resnet50.history['val_accuracy'], label='ResNet50 Val Accuracy', color='blue')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Accuracy')
ax1.set_title('ResNet50 Model Training and Validation Accuracy')
ax1.legend()

# Plotting loss and validation loss for ResNet50 model on the second subplot
ax2.plot(epochs, history_resnet50.history['loss'], label='ResNet50 Loss', color='red')
ax2.plot(epochs, history_resnet50.history['val_loss'], label='ResNet50 Val Loss', color='purple')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Loss')
ax2.set_title('ResNet50 Model Training and Validation Loss')
ax2.legend()
plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
print("\033[1mTraining MobileNetV2 Model\033[0m")
history_mobilenetv2 = model_mobilenetv2.fit(
    train_gen_mobilenetv2,
    validation_data=valid_gen_mobilenetv2,
    epochs=20,
    callbacks=[EarlyStopping(monitor='val_accuracy', min_delta=0, patience=4, mode='auto')]
)
print("\033[1mTraining InceptionV3 Model\033[0m")
history_inceptionv3 = model_inceptionv3.fit(
    train_gen_inceptionv3,
    validation_data=valid_gen_inceptionv3,
    epochs=20,
    callbacks=[EarlyStopping(monitor='val_accuracy', min_delta=0, patience=4, mode='auto')]
)
print("\033[1mTraining VGG16 Model\033[0m")
history_vgg16 = model_vgg16.fit(
    train_gen_vgg16,
    validation_data=valid_gen_vgg16,
    epochs=15,
    callbacks=[EarlyStopping(monitor='val_accuracy', min_delta=0, patience=3, mode='auto')]
)
```

```python
import cv2
import numpy as np
from tensorflow.keras.applications.resnet50 import preprocess_input as preprocess_input_resnet50
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input as preprocess_input_mobilenetv2
from tensorflow.keras.applications.inception_v3 import preprocess_input as preprocess_input_inceptionv3
from tensorflow.keras.applications.vgg16 import preprocess_input as preprocess_input_vgg16

# Predicting and visualizing the results for a sample image
def predict_image(model, img_path):
    img = cv2.imread(img_path)
    if model.name == 'loaded_model_vgg16':
        img = cv2.resize(img, (224, 224))  # Adjusting size for VGG16
        x = np.expand_dims(img, axis=0)
        x = preprocess_input_vgg16(x)
    elif model.name == 'loaded_model_resnet50':
        img = cv2.resize(img, (224, 224))  # Adjusting size for ResNet50
        x = np.expand_dims(img, axis=0)
        x = preprocess_input_resnet50(x)
    elif model.name == 'loaded_model_mobilenetv2':
        img = cv2.resize(img, (224, 224))  # Adjusting size for MobileNetV2
        x = np.expand_dims(img, axis=0)
        x = preprocess_input_mobilenetv2(x)
    elif model.name == 'loaded_model_inceptionv3':
        img = cv2.resize(img, (299, 299))  # Adjusting size for InceptionV3
        x = np.expand_dims(img, axis=0)
        x = preprocess_input_inceptionv3(x)
    else:
        raise ValueError("Unknown model name")
    result = model.predict(x)
    return result
```
